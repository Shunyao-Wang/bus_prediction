```python
import pandas as pd
import numpy as np
import gc
```

```python
df2 = pd.read_csv('../bus_data/stationmap/location_merge_near.csv')
df1 = pd.read_csv('../bus_data/stationmap/train_amap_correct.csv')
```

```python
df1
```

```python
df2
```

```python
a_x = np.ones((len(df2),len(df1)))
a_x.shape
```

```python
ttt = np.array(df1.longitude_amap)
ttt
```

```python
a_x = ttt * a_x
a_x = a_x.T
a_x
```

```python
b_x = np.ones((len(df1),len(df2)))
ttt = np.array(df2.longitude)
b_x = ttt * b_x
b_x
```

## 计算y差值

```python
a_y = np.ones((len(df2),len(df1)))
ttt = np.array(df1.latitude_amap)
```

```python
a_y = a_y * ttt
a_y = a_y.T
a_y
```

```python
b_y = np.ones((len(df1),len(df2)))
ttt = np.array(df2.latitude)
```

```python
b_y = ttt * b_y
b_y
```

```python
distance = (a_x*1000 - b_x*1000)**2 + (a_y*1000 - b_y*1000)**2
distance
```

```python
del [a_x,a_y,b_x,b_y]
gc.collect()
```

## 保存数据

```python
np.save('match.npy',distance)
```

## 读取数据

```python
distance = np.load('match.npy')
distance
```

```python
aaa = []
for i in range(distance.shape[0]):
#for i in range(1000):
    ttt = np.where(distance[i,:] == distance[i,:].min())
    realid = ttt[0][0]
    a = df2[df2.station_id == realid]
    aaa.append(a.index[0])
aaa
```

```python
result = df1.copy()
result['station_id'] = aaa
result
```

```python
result = pd.merge(left=result,right=df2,left_on=['station_id'],right_on=['station_id'],how='left')
result
```

```python
result=result.rename(columns={'longitude':'real_longitude','latitude':'real_latitude'})
result
```

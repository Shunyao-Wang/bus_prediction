```python
import pandas as pd
import numpy as np
```

```python
a = pd.read_csv('../bus_data/stationmap/match_result_merge_near.csv')
a
```

```python
for i in range(1001,1032):
    b = pd.read_csv('../bus_data/traveltime_records/traveltime_record_2017{}.csv'.format(str(i)))
    del b['O_LONGITUDE_START']
    del b['O_LONGITUDE_END']
    del b['O_LATITUDE_START']
    del b['O_LATITUDE_END']
    del b['O_SPEED_START']
    del b['O_SPEED_END']
    del b['AVER_SPEED']
    
    b1 = pd.merge(left=b,right=a[['O_LINENO','O_UP','STATION','station_id']],left_on=['O_LINENO_START','O_UP_START','ARRIV_STATION_START'],right_on=['O_LINENO','O_UP','STATION'],how='left')
    del b1['O_LINENO']
    del b1['O_UP']
    del b1['STATION']
    b1=b1.rename(columns={'station_id':'STATION_ID_START'})
    
    b2 = pd.merge(left=b1,right=a[['O_LINENO','O_UP','STATION','station_id']],left_on=['O_LINENO_END','O_UP_END','ARRIV_STATION_END'],right_on=['O_LINENO','O_UP','STATION'],how='left')
    del b2['O_LINENO']
    del b2['O_UP']
    del b2['STATION']
    b2=b2.rename(columns={'station_id':'STATION_ID_END'})
    
    b2.fillna(-1,inplace=True)
    b2.to_csv('../bus_data/traveltime_records_merge_near/2017{}.csv'.format(str(i)),index=False,float_format='%.0f')
    print('2017{}.csv'.format(str(i)))
```

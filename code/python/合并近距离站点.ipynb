```python
import pandas as pd
import numpy as np
```

```python
df1 = pd.read_csv('../bus_data/stationmap/location_corrected.csv')
df1
```

```python
a_x = np.ones((len(df1),len(df1)))
ttt = np.array(df1.longitude)
a_x = ttt * a_x
a_x = a_x.T
a_x
```

```python
b_x = a_x.T
b_x
```

```python
a_y = np.ones((len(df1),len(df1)))
ttt = np.array(df1.latitude)
a_y = ttt * a_y
b_y = a_y
a_y = b_y.T
display(a_y,b_y)
```

```python
distance = ((a_x - b_x)*1000)**2 + ((a_y - b_y)*1000)**2
distance
```

```python
np.where((distance[50,60:]<1000 ))
```

### *`distance`*取0.9~1左右为80米

```python
df_temp = pd.DataFrame(columns=['station_id','name','longitude','latitude'])
df_temp
```

```python
df2 = df_temp.copy()
appear = set()
a = 0
for i in range(distance.shape[0]):
#for i in range(1000):
    if i in appear:
        #print(i)
        continue
    else:
        appear.add(i)
        ttt = np.where(distance[i,i:] < 0.8)
        if len(ttt[0]) == 1:
            df2 = df2.append(df1.loc[i,:],ignore_index=True)
        else:
            temp = {}
            temp['station_id'] = i
            temp['name'] = '{}'.format(df1.loc[i,'name'])
            lon_temp = df1.loc[i,'longitude']
            lat_temp = df1.loc[i,'latitude']
            for j in range(1,len(ttt[0])):
                temp['name'] = temp['name'] + '/' + '{}'.format(df1.loc[ttt[0][j]+i,'name'])
                lon_temp += df1.loc[ttt[0][j]+i,'longitude']
                lat_temp += df1.loc[ttt[0][j]+i,'latitude']
                appear.add(ttt[0][j]+i)
            temp['longitude'] = lon_temp/(len(ttt[0]))
            temp['latitude'] = lat_temp/(len(ttt[0]))
            df2 = df2.append(temp,ignore_index=True)
            #print(temp)
            #print(i,ttt[0]+i)
            a += 1
print(a)
df2
```

```python
del df2['station_id']
```

```python
df2.to_csv('../bus_data/stationmap/location_merge_near.csv',index_label='station_id')
```

```python
import numpy as np
import pandas as pd
```

```python
filename = '20180803_183645_487611'
df1 = pd.read_csv('../bus_data/result/{}.csv'.format(filename))
df1
```

```python
time_add = df1.pred_timeStamps.str.split(';')
time_add
```

```python
time = []
for j in range(len(time_add)):
    temp = [float(item) for item in time_add[j]]
    for i in range(len(temp)-1,0,-1):
        temp[i] = temp[i] - temp[i-1]
    temp = [str(item) for item in temp]
    temp = ";".join(temp)
    time.append(temp)
time
```

```python
df2 = df1.copy()
df2.pred_timeStamps = time
df2
```

```python
df2.to_csv('../bus_data/result/time/time_{}.csv'.format(filename),index=False)
print('time',filename)
```

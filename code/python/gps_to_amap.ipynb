```python
import pandas as pd
import numpy as np
import requests  # 导入requests
from bs4 import BeautifulSoup  # 导入bs4中的BeautifulSoup
df_train = pd.read_csv('../bus_data/stationmap/station_geo_position_v2.csv')
#df_real = pd.read_csv('../bus_data/stationmap/location_corrected.csv',index_col=0)
df_train
```

```python
df = df_train.copy()
df['longitude_amap'] = np.NaN
df['latitude_amap'] = np.NaN
df
```

```python
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/49.0.2623.221 Safari/537.36 SE 2.X MetaSr 1.0'}
for i in range(len(df)):
#for i in range(5):
    gps_x = df.loc[i,'O_LONGITUDE']
    gps_y = df.loc[i,'O_LATITUDE']
    url = 'https://restapi.amap.com/v3/assistant/coordinate/convert?locations=' + str(gps_x) + ',' + str(gps_y) +\
        '&coordsys=gps&output=xml&key=faf27f6e255729f070ae62bc0f5c6a74'  # 开始的URL地址
    start_html = requests.get(url, headers=headers)
    Soup = BeautifulSoup(start_html.text, 'lxml')
    all_a = Soup.find('locations')
    
    #if all_a == None:
    #    continue
    amap = all_a.get_text()
    amap = amap.split(',')
    df.loc[i,'longitude_amap'] = amap[0]
    df.loc[i,'latitude_amap'] = amap[1]
    if i%100 == 0:
        print('complete {}'.format(i))
```
